```python
from bs4 import BeautifulSoup
import requests
import re
import json

SRC = 'http://www.1000mest.ru/city'
```

```python
response = requests.get(SRC)
soup = BeautifulSoup(response.text, "html.parser")
pages = set()
for link in soup.findAll('a'):
    if re.match(r'city*', link['href']) is not None:
        pages.add('http://www.1000mest.ru/' + link['href'])
```

```python
cities = set()
for page in pages:
    page_response = requests.get(page)
    page_soup = BeautifulSoup(page_response.text, "html.parser")
    for city in page_soup.findAll('td'):
        cities.add(city.text.split('(')[0])
```

```python
with open('cities.json', 'w') as out: 
    json.dump(list(cities), out, ensure_ascii=False)
print(cities)
```
